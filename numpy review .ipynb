```python
import numpy as np 
import matplotlib.pyplot as plt 
from matplotlib import style 
```

```python
style.use('seaborn')
freq,mic1,mic2 = np.loadtxt('microphones.txt',unpack=True)
plt.plot(freq,mic1,'r-',freq,mic2,'b-')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.legend(['Microphone 1','Microphone 2'])
plt.show()
```

```python
x = np.linspace(1,10,1000)
y1 = x**4*np.exp(-2*x)
y2 = (x**2*np.exp(-x)*np.sin(x**2))**2

plt.plot(y1,'r',y2,'b')
plt.legend(['$x**4*exp(-2*x)$','($x**2*np.exp(-x)*np.sin(x**2))**2$'])
plt.title('Graph of The Function')
plt.show()
```

```python
# bisection mehtod for finding the roots in any equatin
tolerance = 1.0e-5
def f(x):
    f = np.exp(x)*np.log(x) - x*x
    return f
a = 8
b = 9
dx = abs(b-a)
while dx > tolerance:
    x = (a+b)/2
    if (f(a)*f(x)) < 0:
        b = x
    else:
        a = x 
        dx = abs(b-a)
print('found the root',x,tolerance)
```
