```python
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib import style 
```

```python
style.use('classic')
x = np.linspace(1,100,100)
y = np.sin(x)
plt.plot(y)
plt.show()
```

```python
x = np.linspace(0,2*np.pi,100)
ya = np.sin(x)
yb = np.cos(x)
plt.plot(x,ya)
plt.plot(x,yb)
plt.grid()
plt.legend(['sin(x)','cos(x)'])
plt.show()
```

```python
def plot_slope(x,y):
    xs = x[1:] - x[:-1]
    ys = y[1:] - y[:-1]
    plt.plot(x[1:],ys/xs)
    
```

```python
x = np.linspace(-3,3,100)
y = np.exp(-x**2)
plt.plot(x,y)
plot_slope(x,y)
plt.show()
```

```python
x = np.linspace(0,2*np.pi,100)
y = x**2 * np.cos(x)
plt.plot(x,y)
plot_slope(x,y)
```
